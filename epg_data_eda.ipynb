```python
import pandas as pd
df = pd.read_csv("epg_data-2021-12-14.csv", delimiter=";")
df.head(40)
```

```python
df.isnull().sum()
```

# Most Occurring starting time and dates

```python
df_air_dates = df[["id","start_dt"]]
df_air_dates[["date","time"]] = df_air_dates["start_dt"].str.split("T", expand=True)
df_air_dates = df_air_dates.drop(["start_dt"],axis=1)
df_air_dates[["year","month", "day"]] = df_air_dates["date"].str.split("-", expand=True)
df_air_dates = df_air_dates.drop(["date"],axis=1)
df_air_dates[["hr","min", "sec"]] = df_air_dates["time"].str.split(":", expand=True)
df_air_dates = df_air_dates.drop(["time","sec"],axis=1)
df_air_dates["start_time"] = 60*df_air_dates["hr"].astype(int) + df_air_dates["min"].astype(int)
df_air_dates = df_air_dates.drop(["hr","min"],axis=1)

df_air_dates["month"].value_counts().head(12).plot(kind='bar',xlabel="month", ylabel="count")
```

```python
df_air_dates["day"].value_counts().head(31).plot(kind='bar',xlabel="day", ylabel="count")
```

# Relation Between Channel and Categories

```python
df_n = df[["titles", "categories", "descriptions.text","channel","diffusion_country"]]
df_n["titles"] = df_n["titles"].dropna().apply(lambda x:literal_eval(x))
df_n["categories"] = df_n["categories"].dropna().apply(lambda x:literal_eval(x))
df_n = df_n.explode("titles")
df_n = df_n.explode("categories")
df_n = df_n.drop_duplicates().reset_index(drop=True)
df_n
```

# Number Of Unique Titles on Top 30 Channels

```python
df_each_unq_title = df_n[["titles", "channel"]].drop_duplicates().reset_index(drop=True)
df_each_unq_title["channel"].value_counts().head(30).plot(kind='bar',xlabel="channel", ylabel="number of unique titles")
```

# Word Count Analysis

```python
#check for duplicates in descriptions
df_n["descriptions.text"].duplicated(keep="first").value_counts()
```

```python
#drop duplicates in descriptions
df_n.drop_duplicates(subset="descriptions.text",keep="first",inplace=True,ignore_index=True)
```

```python
#number of words in each description
df_n['num_desc_words'] = [len(x.split()) for x in df_n['descriptions.text'].tolist()]
df_n
```

```python
#number of words in description
df_n['num_desc_words'].describe()
```

```python
#number of words in description category wise

df_n.groupby(['categories'])['num_desc_words'].describe()
```

```python
#number of words in description channel wise

df_n.groupby(['channel'])['num_desc_words'].describe()
```

```python
#drop duplicates in titles
df_ut = df_n.drop_duplicates(subset="titles",keep="first",ignore_index=True)
df_ut['num_title_words'] = [len(x.split()) for x in df_ut['titles'].tolist()]
df_ut
```

```python
#number of words in titles
df_ut["num_title_words"].describe()
```

```python
#number of words in tiles channel wise
df_ut.groupby(['channel'])['num_title_words'].describe()
```

```python
#number of words in tiles diffusion_country wise
df_ut.groupby(['diffusion_country'])['num_title_words'].describe()
```

```python
#number of words in tiles categories wise

df_ut.groupby(['categories'])['num_title_words'].describe()
```
